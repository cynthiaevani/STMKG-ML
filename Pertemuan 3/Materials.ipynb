#NumPy

```python
import numpy as np
```

```python
# matrix properties
mat = np.arange(15).reshape(3,5)
print("matrix =\n", mat,
      "\nshape:", mat.shape,
      "\ndimension:", mat.ndim,
      "\ntype:", mat.dtype)
```

```python
# array creation
first = [1,2,3,4]
second = [5,6,7,8]
myarray = np.array([first, second])
print(myarray)
```

```python
# array creation
first = [1,2,3,4]
second = [5,6,7,8]
myarray = np.array([first, second], dtype=np.float64)
print(myarray)
```

```python
# array creation (all zeros)
zero = np.zeros((3,4))
print(zero)
```

```python
# array creation (all ones)
one = np.ones((3,4))
print(one)
```

```python
# basic operations
A = np.array([[1,1],[0,1]])
B = np.array([[2,0],[3,4]])
print("element-wise product:\n", A*B,
      "\nmatrix product:\n", A@B,
      "\nsubstraction:\n", A-B)
```

```python
A = np.array([[1,1],[0,1]])
B = np.array([[2,0],[3,4]])
print("addition:\n", np.add(A,B),
      "\nmatrix product:\n", np.dot(A,B))
```

```python
C = np.random.randint(10, size=4)
print("array =", C,
      "\nsum:", C.sum(),
      "\nmin value:", C.min(),
      "\nmin position:", C.argmin(),
      "\nmax value:", C.max(),
      "\nmax position:", C.argmax(),)
```

```python
D = np.arange(4)
print("matrix =", D,
      "\nexponent:\n", np.exp(D),
      "\nsquare-root:\n", np.sqrt(D))
```

```python
# indexing and slicing (one-dimensional array)
A = np.arange(10)
print("A =", A,
      "\nthe first 5 elements:", A[:5],
      "\nthe first 8 elements, skip every 2 elements:", A[:8:2],
      "\nreversed order:", A[::-1])
```

```python
# indexing and slicing (multi-dimensional array)
B = np.arange(12).reshape(3,4)
print("B =\n", B,
      "\neach row in the 1st column:", B[:,0],
      "\neach row in the 3rd column:", B[:,2],
      "\neach column in 2nd and 3rd row:\n", B[1:3,:])
```

```python
df = pd.read_csv("dirtydata.csv")
df.info()
```

```python
np.where(df.isnull())
```

```python
new_df = df.dropna()
new_df.info()
```

```python
new_df = df.fillna(130)
new_df.info()
```

```python
new_df = df
new_df["Calories"].fillna(130, inplace=True)
new_df.info()
```

```python
new_df = df
x = df["Calories"].mean()
new_df["Calories"].fillna(x, inplace=True)
print("mean =", x)
print(new_df.info())
```

```python
new_df = df
x = df["Calories"].median()
new_df["Calories"].fillna(x, inplace=True)
print("median =", x)
print(new_df.info())
```

```python
new_df = df
x = df["Calories"].mode()[0]
new_df["Calories"].fillna(x, inplace=True)
print("mean =", x)
print(new_df.info())
```

```python
new_df["Date"] = pd.to_datetime(new_df["Date"])
print(new_df.to_string())
```

```python
new_df.dropna(subset=["Date"], inplace=True)
print(new_df.to_string())
```

```python
new_df.loc[7, "Duration"] = 45
print(new_df.loc[7])
```

```python
new_df.loc[7, "Duration"] = 450
```

```python
for x in new_df.index:
  if new_df.loc[x, "Duration"] > 120:
    new_df.loc[x, "Duration"] = 120

print(new_df.loc[7])
```

```python
for x in new_df.index:
  if new_df.loc[x, "Duration"] > 120:
    new_df.drop(x, inplace=True)

new_df.head(10)
```

```python
new_df.duplicated()
```

```python
new_df.drop_duplicates(inplace=True)
new_df.duplicated()
```
