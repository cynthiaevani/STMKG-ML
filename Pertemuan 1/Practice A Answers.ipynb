# Simple Calculator
Lengkapilah kode berikut untuk membuat program sederhana yang dapat melakukan operasi penjumlahan, pengurangan, perkalian, pembagian, pangkat, dan akar dua angka integer berdasarkan input pengguna. <br />
Program akan terus meminta masukan pengguna hingga pengguna memasukkan 'exit' sebagai input.

```python
def add(a, b):
  """Fungsi untuk operasi penjumlahan a+b"""
  # YOUR CODE HERE
  c = a + b
  return c

def sub(a, b):
  """Fungsi untuk operasi pengurangan a-b"""
  # YOUR CODE HERE
  c = a - b
  return c

def mul(a, b):
  """Fungsi untuk operasi perkalian a*b"""
  # YOUR CODE HERE
  c = a * b
  return c

def div(a, b):
  """Fungsi untuk operasi pembagian a/b"""
  # YOUR CODE HERE
  c = a / b
  return c

def exp(a, b):
  """Fungsi untuk operasi pangkat a^b"""
  # YOUR CODE HERE
  c = a ** b
  return c

def root(a,b):
  """Fungsi untuk operasi akar bâˆša"""
  # YOUR CODE HERE
  c = a ** (1/b)
  return c

def input_num():
  """Fungsi untuk meminta masukan angka dari pengguna"""
  # HINT: bisa menggunakan fungsi split() untuk multiple inputs in one line dan 
  #       konsep list comprehension
  # YOUR CODE HERE
  a, b = [int(x) for x in input("Masukkan angka: ").split()]
  return a, b
```

```python
def main():
  command_list = ("add", "sub", "mul", "div", "exp", "root", "exit")
  # cetak list perintah dengan format <nomor>.<spasi><perintah><enter>
  # contoh: 1. add
  #         2. sub
  #         ...
  # YOUR CODE HERE
  print("List perintah: ")
  i = 1
  for item in command_list:
    print("{}. {}".format(i, item))
    i += 1

  command = input("Masukkan perintah: ")
  while command != "exit":

    if command == "add":
      a, b = input_num()
      result = add(a,b)
      print("Hasil penjumlahan adalah {}.\n".format(result))
    
    # lengkapi untuk keenam operasi
    # YOUR CODE HERE
    elif command == "sub":
      a, b = input_num()
      result = sub(a,b)
      print("Hasil pengurangan adalah {}.\n".format(result))

    elif command == "mul":
      a, b = input_num()
      result = mul(a,b)
      print("Hasil perkalian adalah {}.\n".format(result))
    
    elif command == "div":
      a, b = input_num()
      result = div(a,b)
      print("Hasil pembagian adalah {}.\n".format(result))

    elif command == "exp":
      a, b = input_num()
      result = exp(a,b)
      print("Hasil pangkat adalah {}.\n".format(result))

    elif command == "root":
      a, b = input_num()
      result = root(a,b)
      print("Hasil akar adalah {}.\n".format(result))
    
    else:
      print("Perintah tidak valid!")
    
    command = input("Masukkan perintah: ")

  print("Exiting... Done.")
```

```python
if __name__ == "__main__":
  main()
```
